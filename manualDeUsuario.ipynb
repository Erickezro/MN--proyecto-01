# Manual de Usuario

## Requisitos del Sistema

Antes de ejecutar el programa, asegúrate de cumplir con los siguientes requisitos y configuraciones en tu sistema:

1. **Instalación de Python**  
   Es fundamental tener instalado Python en tu ordenador. Para verificar si está disponible, utiliza el siguiente comando en tu terminal preferida:
   ```bash
   python --version
   ```
   Si Python no está instalado, puedes descargarlo desde la página oficial: [Descargar Python](https://www.python.org/downloads/).

2. **Instalación de Librerías Necesarias**  
   El programa utiliza varias librerías para su correcto funcionamiento. Instálalas utilizando el gestor de paquetes `pip` con el siguiente comando:  
   ```bash
   pip install nombre_libreria
   ```
   A continuación, se detalla la lista de librerías necesarias y su propósito:
   - **Tkinter**: Creación de la interfaz gráfica.
   - **PIL (Pillow)**: Manejo y visualización de imágenes en la interfaz.
   - **Matplotlib**: Generación de gráficos estáticos y animados.
   - **Scipy**: Realización de cálculos avanzados (por ejemplo, interpolaciones).
   - **Numpy**: Operaciones matemáticas, manejo de vectores y matrices.

   Ejemplo para instalar `Numpy`:
   ```bash
   pip install numpy
   ```

3. **Verificación de Instalación**  
   Para confirmar que todas las librerías están instaladas correctamente, utiliza el comando:
   ```bash
   pip list
   ```
   Esto mostrará una lista de todas las librerías disponibles en tu entorno de Python. Si falta alguna, instálala antes de continuar.

Cuando hayas completado estos pasos, estarás listo para ejecutar el programa.

---

## Ejecución del Programa

Sigue los pasos a continuación para ejecutar el programa y realizar la simulación:

1. **Iniciar el Programa**  
   Al ejecutar el programa, se desplegará una ventana de bienvenida que muestra información general sobre el proyecto y sus autores. En esta ventana, encontrarás un botón titulado **"Ir al ingreso de datos"**. Haz clic en este botón para acceder al formulario de entrada.

2. **Formulario de Datos**  
   En esta sección, ingresa los parámetros necesarios para la simulación, como la velocidad, altura, y demás valores requeridos. El programa validará automáticamente los datos ingresados:
   - Si los datos son correctos, se procederá a la simulación.
   - Si hay errores (por ejemplo, una velocidad negativa), aparecerá un mensaje de advertencia indicando el problema. Podrás corregir los datos y reintentarlo.

3. **Ejecución de la Simulación**  
   Una vez que los datos sean válidos, el programa mostrará una simulación en la que se visualiza el lanzamiento de una bomba hacia un cañón. Dependiendo de los valores ingresados, los resultados variarán.

4. **Resultados y Retroalimentación**  
   Durante y después de la simulación, podrás observar cómo los parámetros afectan el comportamiento de la bomba. Si se produce un impacto en el cañón o el suelo, el programa generará un mensaje indicando las coordenadas exactas del impacto.

---

### Recomendaciones

- Ingresa valores realistas y positivos para obtener simulaciones más precisas.
- Si encuentras problemas, verifica que las librerías estén instaladas correctamente y revisa los datos ingresados.
- Familiarízate con los parámetros antes de iniciar la simulación para aprovechar todas las funcionalidades del programa.
